Objective:  To have a look at how SVMs perform on data generated from the standard linear model

```r
# install.packages("kernlab",repos="http://cran.rstudio.com/" )
library(kernlab)
```

```r
set.seed(123)

nobs <- 50

e <- rnorm(nobs,0,10)
x <- seq(1,100, length.out=nobs)
intercept <- 5
b <- 2

y <- intercept + b * x + e
df <- data.frame(y,x,e)


```

```r
svm_model <- ksvm(y ~ x, data = df)
library(ggplot2)
ggplot(data=df, aes(x=x)) +
     geom_point(aes(y=y), alpha=0.2) + 
     scale_colour_hue(h = c(180,0)) + 
     geom_line(aes(y=predict(svm_model,df)))
```

```r
nobs <- 100

e <- rnorm(nobs,0,10)
x <- seq(1,10, length.out=nobs)
intercept <- 5
b <- 4
y <- intercept + b*sin(x) + e

df_sin <- data.frame(y, x, e)

svm_model <- ksvm(y ~ x, data = df_sin,kernel='vanilladot')
library(ggplot2)
ggplot(data=df_sin, aes(x=x)) +
     geom_point(aes(y=y), alpha=0.2) + 
     scale_colour_hue(h = c(180,0)) + 
     geom_line(aes(y=predict(svm_model,df_sin)))
```
